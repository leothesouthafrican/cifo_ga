```python
#Importing libraries
import pandas as pd
import os
```

```python
#Import different dfs

def single_alg_dataframe_importer(path):

    list_of_dfs = []
    for i in os.listdir(path):
        if i != ".DS_Store":
            temp_df = pd.read_excel(path + i)
            temp_df = temp_df.iloc[0:199,1:-1]
            list_of_dfs.append(temp_df)
    
    return list_of_dfs




single_alg_dataframe_importer("./results/spec_2/")[0]

```

```python
def avg_fitness_per_gen(algo_1_path):

    for df in algo_1_path:
        total_fitness = 0
        total_fitness += df.iloc[0,0]
        average_fitness = total_fitness/len(algo_1_path)
    return average_fitness

avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_1/"))
```
