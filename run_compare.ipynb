```python
#Importing libraries
import pandas as pd
import os
```

```python
#Import different dfs

def single_alg_dataframe_importer(path):

    list_of_dfs = []
    for i in os.listdir(path):
        if i != ".DS_Store":
            temp_df = pd.read_excel(path + i)
            temp_df = temp_df.iloc[0:199,1:-1]
            list_of_dfs.append(temp_df)
    
    return list_of_dfs

```

```python
def avg_fitness_per_gen(algo_1_path):
    full_column = []
    for generation in range(len(algo_1_path[0])):
        
        total_fitness = 0
        for df in algo_1_path:
            item_to_add = df.iloc[generation:generation+1,0:1]["best_fitness"]
            total_fitness += item_to_add.values[0]
        average_fitness = total_fitness/(len(algo_1_path))
        full_column.append(average_fitness)

    return pd.Series(full_column)
```

```python
new_df = pd.DataFrame()

new_df["spec_0_1"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_0 (fps,pmx,inv)/fitness_1/"))
new_df["spec_0_2"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_0 (fps,pmx,inv)/fitness_2/"))

new_df["spec_1_1"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_1 (tour,pmx,inv)/fitness_1/"))
new_df["spec_1_2"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_1 (tour,pmx,inv)/fitness_2/"))

new_df["spec_2_1"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_2 (fps,geo,swap)/fitness_1/"))
new_df["spec_2_2"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_2 (fps,geo,swap)/fitness_2/"))

new_df["spec_3_1"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_3 (tour,ari,geo)/fitness_1/"))
new_df["spec_3_2"] = avg_fitness_per_gen(single_alg_dataframe_importer("./results/spec_3 (tour,ari,geo)/fitness_2/"))
```
