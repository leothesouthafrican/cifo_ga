```python
#Importing libraries
import pandas as pd
import plotly.express as px
import os
import openpyxl
```

```python
#Import different dfs

def single_alg_dataframe_importer(path):

    list_of_dfs = []
    for i in os.listdir(path):
        if i != ".DS_Store":
            temp_df = pd.read_excel(path + i)
            temp_df = temp_df.iloc[0:199,1:-1]
            list_of_dfs.append(temp_df)
    
    return list_of_dfs
```

```python
def avg_metric_per_gen(algo_1_path, column):
    full_column = []
    for generation in range(len(algo_1_path[0])):
        
        total_metric = 0
        for df in algo_1_path:
            item_to_add = df.iloc[generation:generation+1,0:8][column]
            total_metric += item_to_add.values[0]
        average_metric = total_metric/(len(algo_1_path))
            
        full_column.append(average_metric)

    return pd.Series(full_column)
```

```python
fitness_df = pd.DataFrame()

fitness_df["spec_0_1"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_0 (fps,pmx,inv)/fitness_1/"),"best_fitness")
fitness_df["spec_0_2"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_0 (fps,pmx,inv)/fitness_2/"),"best_fitness")

fitness_df["spec_1_1"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_1 (tour,pmx,inv)/fitness_1/"),"best_fitness")
fitness_df["spec_1_2"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_1 (tour,pmx,inv)/fitness_2/"),"best_fitness")

fitness_df["spec_2_1"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_2 (fps,geo,swap)/fitness_1/"),"best_fitness")
fitness_df["spec_2_2"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_2 (fps,geo,swap)/fitness_2/"),"best_fitness")

fitness_df["spec_3_1"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_3 (tour,ari,geo)/fitness_1/"),"best_fitness")
fitness_df["spec_3_2"] = avg_metric_per_gen(single_alg_dataframe_importer("./results/spec_3 (tour,ari,geo)/fitness_2/"),"best_fitness")
```

### Developing Analysis for Different Sheet Format

```python
class Analysis:
    def __init__(self, path, metric, population):
        self.path = path
        self.metric = metric
        self.population = population
    #Function that computes average value per epoch across n runs
    def average_value_per_epoch(path, metric, population):
        
        #List of all files in the directory
        file_list = os.listdir(path)
        #Iterate through all excels
        for file in file_list:
            for generation in range(population):
                #Storing all sheets in each excel file
                wb = openpyxl.load_workbook(path + file)
                #Getting number of sheets contained within excel
                full_column = []
                total_metric = 0
                for sheet in wb.sheetnames:
                    temp_sheet = pd.read_excel(path + file, sheet_name= sheet) 
                
                    value_to_add = int(temp_sheet.iloc[generation,metric])
                    total_metric += value_to_add
                average_metric = total_metric/len(wb.sheetnames)

                print(f"Total for Sheet: {average_metric}")
```
